```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load the housing dataset
housing = fetch_california_housing()

X = pd.DataFrame(housing.data, columns=housing.feature_names) 
y = pd.Series(housing.target, name='med_house_value')
```

```python
# Display the first five rows of the dataset
X.head()
```

Features explained from https://scikit-learn.org/stable/datasets/real_world.html#california-housing-dataset (Note to myself).

The target variable is the median house value for California districts, expressed in hundreds of thousands of dollars ($100,000).

MedInc - median income in block group.
HouseAge - median house age in block group.
AveRooms - average number of rooms per household.
AveBedrms - average number of bedrooms per household.
Population - block group population.
AveOccup - average number of household members.
Latitude - block group latitude.
Longitude - block group longitude.

```python
# Print the feature names and check for missing values
X.info()
```

```python
# Generate summary statistics (mean, min, max, etc.)
X.describe()
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   random_state=42,
                                                   test_size=0.2)

# Train a linear regression model on the unscaled data using sklearn.linear_model.LinearRegression
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = lin_reg.predict(X_test)
```

```python
# Evaluate model performance using MSE, RMSE, R^2
from sklearn.metrics import mean_squared_error, root_mean_squared_error, r2_score

# Evaluate model performance
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)

r2 = r2_score(y_test, y_pred)

print("Unscaled Data Model:")
print(f"Mean Squared Error: {mse:.2f}")
# On average the model fits within two points of predicting
print(f"Root Squared Error: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
print(pd.Series(lin_reg.coef_,
                index=X.columns))
```

The R^2 tells us that about 58% of the variance in our dependent variable can be explained using our model. The model is capturing more than half of what we want it to but failing to pickup on a large portion of the data. The average number bedrooms feature has the largest impact on the outcome variable. For a 1 unit increase in average number of bedrooms the median house value for Californian districts increases by roughly 0.78. The Longitude and Latituted have the largest negative impact on the predictions. A 1 unit increase in longitude decreases the median house value by roughly 0.43. The root squared error tells us that predictions are off by about 0.75 units on average. 



```python
# Select three features from the dataset to build a simplified model. Explain your choice. (10 pts)


X = pd.DataFrame(housing.data, columns=housing.feature_names)

# Now select 3 features
X_select = X[['MedInc', 'AveBedrms', 'AveRooms']]

y_select = pd.Series(housing.target, name='med_house_value')
```

Explanation: I selected the median income column, the average bedrooms column, and the average rooms column because they had the larges coefficient weights in the previous model excluding the location data. So, I figured that these columns would provide better insights without the added noise from the other columns.

```python
X_select_train, X_select_test, y_select_train, y_select_test = train_test_split(X_select, y_select,
                                                   random_state=42,
                                                   test_size=0.2)

# Train a linear regression model on the unscaled data using sklearn.linear_model.LinearRegression
from sklearn.linear_model import LinearRegression

lin_reg_select = LinearRegression()
lin_reg_select.fit(X_select_train, y_select_train)

# Make predictions on the test set
y_pred_select = lin_reg_select.predict(X_select_test)
```

```python
# Evaluate model performance using MSE, RMSE, R^2
mse = mean_squared_error(y_select_test, y_pred_select)
rmse = root_mean_squared_error(y_test, y_pred_select)

r2 = r2_score(y_select_test, y_pred_select)

print("Unscaled Data Model:")
print(f"Mean Squared Error: {mse:.2f}")
# On average the model fits within two points of predicting
print(f"Root Squared Error: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

Compared to the full model this model performed slightly worse. It was only able to explain 48% of the variance in the dependent variable compared to 58% in the previous model. Also, on average this model was 0.82 off from predicting the actual median house value for Californian districts which was 0.07 more than the previous model.

```python
print(pd.Series(lin_reg_select.coef_,
                index=X_select.columns))
```

In practice, a simplified model may be preferred to prevent overfitting. I don't believe the previous model was overfitting but if it was selecting only some of the features in the dataset can reduce the problem of overfitting. By doing this, our train accuracy may decrease but our test accuracy would likely increase.
