```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load the housing dataset
housing = fetch_california_housing()

X = pd.DataFrame(housing.data, columns=housing.feature_names) 
y = pd.Series(housing.target, name='med_house_value')
```

```python
# Display the first five rows of the dataset
X.head()
```

Features explained from https://scikit-learn.org/stable/datasets/real_world.html#california-housing-dataset (Note to myself).

The target variable is the median house value for California districts, expressed in hundreds of thousands of dollars ($100,000).

MedInc - median income in block group.
HouseAge - median house age in block group.
AveRooms - average number of rooms per household.
AveBedrms - average number of bedrooms per household.
Population - block group population.
AveOccup - average number of household members.
Latitude - block group latitude.
Longitude - block group longitude.

```python
# Print the feature names and check for missing values
X.info()
```

```python
# Generate summary statistics (mean, min, max, etc.)
X.describe()
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   random_state=42,
                                                   test_size=0.2)

# Train a linear regression model on the unscaled data using sklearn.linear_model.LinearRegression
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = lin_reg.predict(X_test)
```

```python
# Evaluate model performance using MSE, RMSE, R^2
from sklearn.metrics import mean_squared_error, root_mean_squared_error, r2_score

# Evaluate model performance
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)

r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
# On average the model fits within two points of predicting
print(f"Root Squared Error: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
print(pd.Series(lin_reg.coef_,
                index=X.columns))
```

The R^2 score tells us that the model explains about 58% of the variance in median house value across California districts. This means the model is capturing more than half of what affects housing prices, but fails to explain a large portion of variation still. The median income has the strongest positive effect on home value. The Longitude and Latitude have the largest negative impacts on the home value. A 1-unit change in each feature affects median house value by its coefficient amount in 100,000s of dollars, while holding all other features constant. The root mean squared error (RMSE) shows the model's predictions are off by roughly 75,000 dollars on average. This is not great but it is not horrible for a base model that can be improved upon.



```python
# Select three features from the dataset to build a simplified model. Explain your choice. (10 pts)


X = pd.DataFrame(housing.data, columns=housing.feature_names)

# Now select 3 features
X_select = X[['MedInc', 'AveBedrms', 'AveRooms']]

y_select = pd.Series(housing.target, name='med_house_value')
```

I selected the median income, average bedrooms, and average rooms columns because they had some of the largest coefficient magnitudes in the full model, excluding the location variables. So, I figured that these columns would provide better insights without the added noise from the other columns. Also, this model will be more simple and interpretable due to less features.

```python
X_select_train, X_select_test, y_select_train, y_select_test = train_test_split(X_select, y_select,
                                                   random_state=42,
                                                   test_size=0.2)

# Train a linear regression model on the unscaled data using sklearn.linear_model.LinearRegression
from sklearn.linear_model import LinearRegression

lin_reg_select = LinearRegression()
lin_reg_select.fit(X_select_train, y_select_train)

# Make predictions on the test set
y_pred_select = lin_reg_select.predict(X_select_test)
```

```python
# Evaluate model performance using MSE, RMSE, R^2
mse_sel = mean_squared_error(y_select_test, y_pred_select)
rmse_sel = root_mean_squared_error(y_select_test, y_pred_select)

r2_sel = r2_score(y_select_test, y_pred_select)

print("Three Feature Data Model:")
print(f"Mean Squared Error: {mse_sel:.2f}")
# On average the model fits within two points of predicting
print(f"Root Squared Error: {rmse_sel:.2f}")
print(f"R² Score: {r2_sel:.2f}")
```

Compared to the full model the simple model performed slightly worse. It was only able to explain 48% of the variance compared to 58% in the full model. Also, on average this model was 82,000 dollars off from predicting the actual median house value for Californian districts which was $7,000 more than the previous model. Because this model is more simple it loses some of its predictive power gained from the entire feature set. However, this model with only three variables is much easier to interpret.

```python
print(pd.Series(lin_reg_select.coef_,
                index=X_select.columns))
```

In practice, a simplified model may be preferred to prevent overfitting, or for better interpretability. The previous model was not showing signs of overfitting but, if it was, selecting only some of the features in the dataset can reduce the problem of overfitting. Additionally, a simplified model may be useful in situations where interpretability is more important than accuracy. A quick baseline from the best predictors is the best use case for a simple model. Otherwise, unless it overfits, a full or larger model is preferred.
